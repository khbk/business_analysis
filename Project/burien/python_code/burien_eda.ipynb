```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
burien = pd.read_csv('../data/burien.csv')
```

```python
pd.set_option('display.max_columns', None)
burien.head()
```

```python
burien.columns
```

```python
burien = burien.drop(columns = ['Unnamed: 0', '(Revenue Report)', 
        '잡화 비율', '카드/총매출', 'Unnamed: 17', 'Unnamed: 18', 
        '주', '주별 평균 매출', '평균 담배 갑 ', '평균 담배보루', 
        '평균 그로서리', '평균 잡화', '평균 베입', '평균 크레이톰', 
        '평균 토바코','평균 맥주', '누락', '주별 총 누락', 
        'Unnamed: 31', 'Total', 'Cig', 'Groc', 'Misc',
       'Tob', 'Beer', 'Unnamed: 38', 'Unnamed: 39'])
```

```python
burien.head()
```

```python
burien.rename(columns={'이번 달로 가기':'date', 
                       '총매출':'total', '후카':'hookah', 
                        '메인포스':'main', '카드': 'card'},
             inplace = True)
```

```python
burien.columns = [col.replace(' ', '_').lower() 
                  for col in burien.columns]
```

```python
# Delete rows that contain null values for all
burien.dropna(thresh=10, inplace=True)
```

```python
# We would need to take the '$' out and make them into floats.
for i in burien.columns:
    burien[i] = burien[i].str.replace('$','')
```

```python
burien.head()
```

```python
burien = burien.fillna(0)
```

```python
for i in burien.columns:
    burien[i] = burien[i].str.replace(',','')
```

```python
# Set 'date' column to relate to time series data.
burien['date'] = pd.to_datetime(burien['date'])
```

```python
burien.set_index('date', inplace=True)
```

```python
burien.head()
```

```python
burien = burien.drop(burien.index[burien['total'] == 'C'])
```

```python
burien.replace({'(전기아웃)': '0'}, inplace=True)
```

```python
burien.replace({'p.w out': '0'}, inplace=True)
```

```python
for i in burien.columns:
    burien[[i]] = burien[[i]].astype(float)
```

```python
burien = burien.fillna(0)
```

```python
burien['misc_total'] = burien['misc'] + burien['kratom'] + burien['hookah']
```

```python
burien = burien.drop(columns=['misc', 'kratom', 'hookah'])
```
