```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
up_time = pd.read_csv('../data/up_df.csv')
up_time
```

```python
up_time['date'] = pd.to_datetime(up_time['date'])
```

```python
up_time.set_index('date', inplace = True)
```

```python
up_time.resample('M').mean()
```

```python
plt.figure(figsize=(12,8))
plt.title('Monthly Total', fontsize=16)
plt.ylabel('USD', fontsize=16)
plt.xlabel('Time', fontsize=16)
plt.plot(up_time['total'])
plt.plot(up_time['total'].rolling(30).mean());
```

```python
fig, axes = plt.subplots(nrows=9, figsize=(15,12))
axes[0].plot(up_time['pack'])
axes[1].plot(up_time['carton'])
axes[2].plot(up_time['grcy'])
axes[3].plot(up_time['grcy_nt'])
axes[4].plot(up_time['vape'])
axes[5].plot(up_time['tobacco'])
axes[6].plot(up_time['beer'])
axes[7].plot(up_time['cigar'])
axes[8].plot(up_time['misc_total']);
```
